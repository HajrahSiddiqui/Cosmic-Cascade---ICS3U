```python
#https://www.techwithtim.net/tutorials/game-development-with-python/tetris-pygame/tutorial-1
#https://www.activestate.com/blog/how-to-use-pygame-for-game-development/
#https://www.makeuseof.com/start-menu-and-game-over-screen-with-pygame/
#https://stackoverflow.com/questions/23982907/how-to-center-text-in-pygame
#https://www.geeksforgeeks.org/creating-start-menu-in-pygame/
#sites so far (will add more later)
#https://www.youtube.com/watch?v=7kGNs5R-AM8
#https://data-flair.training/blogs/python-tetris-game-pygame/
#https://stackoverflow.com/questions/33963361/how-to-make-a-grid-in-pygame
#https://www.mygreatlearning.com/blog/python-main/
#https://www.geeksforgeeks.org/python-drawing-different-shapes-on-pygame-window/ 
#https://data-flair.training/blogs/python-tetris-game-pygame/#google_vignette
#https://stackoverflow.com/questions/32959705/player-does-not-move 
#https://excelatfinance.com/xlf/xlf-colors-1.php
#https://www.techwithtim.net/tutorials/game-development-with-python/tetris-pygame/tutorial-1 
#https://stackoverflow.com/questions/32736498/initialize-class-attribute-with-a-method
#https://stackoverflow.com/questions/52104513/python-pygame-class-init-question-not-drawing 
#https://www.programiz.com/article/python-self-why
#https://docs.python.org/3/tutorial/datastructures.html#list-comprehensions 
#https://www.pygame.org/docs/
#https://www.geeksforgeeks.org/pygame-drawing-objects-and-shapes/
#https://stackoverflow.com/questions/66526297/python-how-to-type-anotate-a-method-that-returns-self
#https://www.makeuseof.com/design-and-create-levels-in-pygame/
#https://stackoverflow.com/questions/60468211/how-do-i-draw-text-onto-the-screen-in-pygame
#https://stackoverflow.com/questions/33061063/how-to-make-another-level-in-python-platform-game 
#https://gist.github.com/timurbakibayev/1f683d34487362b0f36280989c80960c
#https://medium.com/wiki-flood/python-tetris-game-build-drop-repeat-ddaf0ed6804c 
#https://stackoverflow.com/questions/19498987/how-to-make-a-sprites-speed-increase-after-a-certain-score-is-achieved-in-pygame
#https://thecodingfun.com/2021/11/09/use-pygame-to-create-tetris-game-an-updated-version-part-2/

#SCHEDULE:

#First make the box in which the program will start/appear
#MAKE THE SHAPES
#Use the matrix idea of making one and then making designs within that
#Program each shape so that there are rotations of each (me and akansh do 3 separate and then the final one will do together in class)
#Program so that the shape can move right left up down (each of us remember)
#Make boundaries for it
#Use the random module to then start making the shapes appear on the screen at random
#Code so that if a row has 10 in the row then it disappears and it gives 10 points to the user
#Add a quit button, a timer of 30 seconds per shape, a background
#The end
#This code from the W to the clock.tick() is from the youtube link https://www.youtube.com/watch?v=7kGNs5R-AM8 

import pygame
import random

#importing the sound
pygame.init()
my_sound = pygame.mixer.Sound('/Users/hajrahsiddiqui/Documents/cosmic cascade sounds/youtube_cat9TbAwRWU_audio.mp3')
my_sound.play()

W = 9
H = 17
TILE = 45
DISPLAY_WIDTH = 200
game_width = W * TILE + DISPLAY_WIDTH
game_height = H * TILE

#Defining the shapes based on their numbers
shapes = [
    [[13, 9, 5, 1], [7, 6, 5, 4]], #Shape I - Akansh
    [[10, 6, 2, 1], [9, 5, 6, 7], [11, 10, 6, 2], [3, 5, 6, 7]], #Shape J - Akansh
    [[6, 5, 4, 1], [9, 5, 1, 4], [9, 6, 5, 4], [9, 5, 1, 6]], #Shape T - Akansh
    [[1, 2, 4, 5], [1, 5, 6, 10]], #Shape S - Hajrah
    [[0, 1, 5, 6], [1, 4, 5, 8]], #Shape Z - Hajrah
    [[2, 6, 10, 11], [7, 9, 10, 11], [1, 2, 6, 10], [9, 10, 11, 13]], #Shape L - Hajrah
    [[0, 1, 2, 3], [0, 4, 8, 12], [12, 13, 14, 15], [3, 7, 11, 15]] #Shape O - Hajrah
]

shape_colors = [
    (128, 0, 128), #Purple
    (255, 0, 255), #Bright pink
    (0, 0, 255), #Blue
    (0, 0, 128), #Navy
    (102, 0, 102), #Dark purple
    (51, 51, 153), #Dark blue
]


class Tile: #Defining the block in the game & displaying the shapes on the grid
    def __init__(self, x, y, tile_type): #Website from above - this is initializing a new tile with a specific x & y position 
        self.x = x #Website from above - sets the x coordinate of the tile on the grid 
        self.y = y #Website from above - sets the y coordinate of the tile on the grid 
        self.type = tile_type #Website from above - this sets the type of the tile, detecting the shapes 
        self.color = tile_type #Website from above - this sets the color of the tile, detecting the shapes  
        self.rotation = 0 #Website from above - this initializng the rotation of the tile to zero 

    def recent_shape(self): #Website from above - this returns the current shape based on its type  
        return shapes[self.type][self.rotation] #Website from above - this gets the current shape based on its design 

    def shape_rotation(self): #Website from above - this rotates the current rotation to the next rotation state 
        self.rotation = (self.rotation + 1) % len(shapes[self.type]) #Website from above - this increments the rotation state within the shape design 


class Game: #Webstie from above - this represents the components of the tetris game 
    def __init__(self): #Website from above - this initializes the the game as it sets up the display and game field  
        pygame.init() #Website from above - this initializes all of the imported pygame modules 
        self.screen_width = game_width #Website from above - this sets the width of the game screen   
        self.screen_height = game_height #Website from above - this sets the height of the game screen  
        self.screen = pygame.display.set_mode((self.screen_width, self.screen_height)) #Website from above - this creates a window with a seperate width and height 
        self.field = [[0 for y in range(W)] for x in range(H)] #GPT - this creates a 2D list representing the game field with their dimensions
        self.clock = pygame.time.Clock() #Website from above - this initalizes a clock object to control the frame rate 
        self.score = 0 #Website from above - this initalizes the score to zero at the start of the game  
        self.gameover = False #Website from above - this indicates whether the game is over or not 
        self.current_tile = self.create_tile() #Website from above - this creates the current tile for the game 
        self.next_tile = self.create_tile() #Website from above - this creates the next tile for the game 
        self.font = pygame.font.SysFont('Calibri', 35) #Website from above - this sets the font for the text in the game
        self.fall_time = 0 #Website from above - this initializes the fall time to zero  
        self.fall_speed = 100 #Website from above - this initializes the falling speed 
        self.start_game = False #Website from above - this tracks whether the game has started or not 

    def reset_game(self): #GPT - this resets the game state to its initial state 
        self.field = [[0 for y in range(W)] for x in range(H)] #GPT - this creates a 2D list representing the game field with their dimensions 
        self.score = 0 #Website from above - this initalizes the score to zero at the start of the game   
        self.gameover = False #Website from above - this indicates whether the game is over or not 
        self.current_tile = self.create_tile() #Website from above - this creates the current tile for the game 
        self.next_tile = self.create_tile() #Website from above - this creates the next tile for the game 
        self.fall_time = 0 #Website from above - this initializes the fall time to zero   
        self.start_game = False #Website from above - this tracks whether the game has started or not 

    def create_tile(self): #Website from above - this creates and returns a new tile at the top of the board 
        return Tile(5, 0, random.randint(0, len(shapes) - 1) % len(shape_colors)) #Website from above - this returns a new tile with a random shape or object 

    def draw_grid(self): #Website from above - this draws the grid lines of the game 
        for x in range(W): #Website from above - this iterates over the width of the grid 
            for y in range(H): #Website from above - this iterates over the height of the grid 
                rect = pygame.Rect(x * TILE, y * TILE, TILE, TILE) #Website from above - this creates a rectangle for each tile in the grid 
                pygame.draw.rect(self.screen, (40, 40, 40), rect, 1) #Website from above - this draws a rectangle on the screen with a boarder color 

    def intersect(self): #Website from above - this checks whether the current tile intersects with the boundries of the game field 
        intersection = False #Website from above - this initializes the intersection as false 
        for y in range(4): #Website from above - this loops through the rows of the current tile 
            for x in range(4): #Website from above - this loops through the colomns of the current tile  
                if y * 4 + x in self.current_tile.recent_shape(): #Website from above - this checks whether the current shape intersects with the field 
                    if y + self.current_tile.y > H - 1: #Webite from above - this checks whether the tile has reached the bottom of the grid
                        return True 
                    if x + self.current_tile.x > W - 1: #Webite from above - this sets a condition where the tile has to reach the right edge of the grid
                        return True
                    if x + self.current_tile.x < 0: #Website from above - this sets a condition where the tile has to reach the left edge of the grid
                        return True
                    if self.field[y + self.current_tile.y][x + self.current_tile.x] > 0: #Website from above - this sets a condition where the tile has to intersect with a non empty cell in the field 
                        return True
        return False

    def break_lines(self): #Website from above - this checks for complete lines and then removes them, increasing the score later
        lines = 0 #Website from above - this initializes a variable to keep track of the number of lines that were removed 
        for y in range(1, H): #Website from above - this iterates over each row in the field 
            empty_cells = 0 #Website from above - this initializes a variable to count the empty cells in the row 
            for x in range(W): #Website from above - this iterates through each colomn in the row
                if self.field[y][x] == 0: #Website from above - this checks if the cell in the current row is empty or not
                    empty_cells += 1 #Website from above - this increments the count of empty cells in the current row 
            if empty_cells == 0: #Website from above - this checks if the current row has no empty cells
                lines += 1 #Website from above - this increments the number of lines cleared 
                for y1 in range(y, 1, -1): #Website from above - this shifts the lines above the cleared lines
                    for x in range(W): #Website frome above - this continues to shift the lines 
                        self.field[y1][x] = self.field[y1 - 1][x] #Website from above - this moves the contents of the current row to the row above it 
        self.score += lines * 100 #Website from above - this increases the score by the number of lines cleared by 100 

    def draw_tile(self, tile): #Website from above - this draws the current falling tile on the screen 
        for y in range(4): #Website from above - this iterates over the rows of the following tiles
            for x in range(4): #Website from above - this iterates over the colomns of the following tiles 
                if y * 4 + x in tile.recent_shape(): #Website from above - this checks if the x or y coordinate has a free tile or not
                    current_x = tile.x + x #Website from above - this calculates the current x coordinate on the grid
                    current_y = tile.y + y #Wesite from above - this calculates the current y coordinate on the grid 
                    if current_y >= 0: #Website from above - this checks whether the current y coordinate on the grid is visable in the grid
                        pygame.draw.rect(self.screen, shape_colors[tile.color], (current_x * TILE, current_y * TILE, TILE, TILE)) #Website from above - this draws the tile on the screen if the y coordinate is visable in the grid

    def draw_next_tile(self): #Website from above - this draws the preview of the next shape on the screen  
        label = self.font.render("Next Shape", True, (255, 255, 255)) #Website from above - this renders the label for the next shape preview  
        self.screen.blit(label, (game_width - DISPLAY_WIDTH + 20, 20)) #Website from above - this displays the label for the next shape preview on the screen
        format = self.next_tile.recent_shape() #Website from above - this gets the current shape format for the next tile
        for y in range(4): #Website from above - this loops through each row 
            for x in range(4): #Website from above - this loops through each coloumn 
                if y * 4 + x in format: #Website from above - this checks if the current cell is part of the shape 
                    current_x = W * TILE + 50 + x * TILE #Website from above - this calculates the x coordinate of the current cell in the Next Shape 
                    current_y = 80 + y * TILE #Website from above - this calculates the y coordinate of the current cell in the Next Shape 
                    pygame.draw.rect(self.screen, shape_colors[self.next_tile.color], (current_x, current_y, TILE, TILE)) #Website from above - this draws a rectangle representing the Next Shape 

    def draw_field(self): #GPT - this draws the tiles that have already been fallen 
        for y in range(H): #Website from above - this iterates through each row in the field  
            for x in range(W): #Website from above - this iterates through each column in the field  
                if self.field[y][x]: #Website from above - this checks if the cell is occupied or not 
                    pygame.draw.rect(self.screen, shape_colors[self.field[y][x] - 1], (x * TILE, y * TILE, TILE, TILE)) #Website from above - this draws the occupied with the shape color 

    def shape_going_down(self): #Website from above - this defines the shapes that are going down
        self.current_tile.y += 1 #Website from above - this moves the current tile down by one  
        if self.intersect(): #Website from above - this checks if the current tile intersects with any other tile 
            self.current_tile.y -= 1 #Website from above - this moves the current tile up by one  
            self.freeze() #Website from above - this locks the current tile in place  
        
    def freeze(self): #Website from above - this defines how the shapes get locked into their places on their grid
        for y in range(4): #Website from above - this loops through each row 
            for x in range(4): #Website from above - this loops through each column 
                if y * 4 + x in self.current_tile.recent_shape(): #Website from above - this checks if the current position is part of the recent shape 
                    self.field[y + self.current_tile.y][x + self.current_tile.x] = self.current_tile.color + 1 #Website from above - this sets the field's cell at the current tile's position 
        self.break_lines() #Website from above - this calls the method to check and break any completed lines in the grid
        self.current_tile = self.next_tile #Website from above - this sets the current tile to the next tile  
        self.next_tile = self.create_tile() #Website from above - this geneartes a new tile to be used as the next tile in the game 
        if self.intersect(): #Website from above - this checks if the current tile intersects with any other tile  
            self.gameover = True #Website from above - this sets the gameover to condition to true  
    
    def go_horizontal(self, delta): #Website from above - this defines the method to move the current tile horizontally
        x = self.current_tile.x #Website from above - this stores the current x coordinate of the current tile  
        self.current_tile.x += delta #Website from above - this updates the x coordinate of the current tile  
        if self.intersect(): #Website from above - this checks if the current tile intersects with any other tile
            self.current_tile.x = x #Website from above - this resets the x coordinate of the current tile to its previous value

    def rotate(self): #Website from above - this defines the method to rotate the current tile 
        rotation = self.current_tile.rotation #Website from above - this stores the current rottion state of the current tile
        self.current_tile.shape_rotation() #Website from above - this rotates the shape of the current tile 
        if self.intersect(): #Website from above - this checks if the current tile intersects with any other tile 
            self.current_tile.rotation = rotation #Website from above - this resets the rotation state of the current tile to its previous value 

    def handle_events(self): #Website from above - this defines the possible events that a player could move or do 
        for event in pygame.event.get(): #Website from above - this itertes thorugh all pygame events  
            if event.type == pygame.QUIT: #Website from above - this checks if the event type is a quit event 
                pygame.quit() #Website from above - this quits the pygame module
                quit() #Website from above - this exits the program 
            if event.type == pygame.KEYDOWN: #Website from above - this checks if the event is a key press event  
                if event.key == pygame.K_UP: #Website from above - this checks if the event is a key pressed up arrow
                    self.rotate() #Website from above - this rotates the current tile clockwise 
                if event.key == pygame.K_DOWN: #Website from above - this checks if the event is a key pressed down arrow 
                    self.shape_going_down() #Website from above - this makes the current tile move downwards  
                if event.key == pygame.K_RIGHT: #Website from above - this checks if the key pressed if the right arrow key  
                    self.go_horizontal(1) #Website from above - this moves the current tile on step to the right 
                if event.key == pygame.K_LEFT: #Website from above - this checks if the key pressed is the left arrow key 
                    self.go_horizontal(-1) #Website from above - this moves the current tile one step to the left
                if event.key == pygame.K_SPACE: #Website from above - this checks if the key pressed if the spacebar key 
                    while not self.intersect(): #Website from above - this continues the looping until the current tile intersects with any other tile 
                        self.current_tile.y += 1 #Website from above - this moves the current tile one step downwards  
                    self.current_tile.y -= 1 #Website from above - this  moves the current tile one step upwards 
                    self.freeze() #Website from above - this locks the current tile in place  

    def display_screen(self): #Website from above - this defines the display screen 
        
        background_image = pygame.image.load('/Users/hajrahsiddiqui/Documents/cosmic cascade sounds/no.png').convert_alpha()
        #background_image = pygame.image.load("/Users/akanshshah/Downloads/PHOTO-2024-05-28-16-27-05-6.jpg")
        self.screen.blit(background_image, (0, 0)) 
        
        #self.screen.fill((0, 0, 0)) #Website from above - this fills the screen with a black color 
        font = pygame.font.SysFont('Calibri', 1) #Website from above - this creates a font object with Calibri 
        title = font.render('Tetris Game!', True, (255, 255, 255)) #Website from above - this renders the text using the font with white color
        start_button = font.render('Click to start!', True, (255, 255, 255)) #Website from above - this renders the text using the font with white color 
        self.screen.blit(title, (self.screen.get_width() / 2 - title.get_width() / 2, self.screen.get_height() / 2 - title.get_height() / 2)) #Website from above - this displays the title text centered on the screen 
        self.screen.blit(start_button, (self.screen.get_width() / 2 - start_button.get_width() / 2, self.screen.get_height() / 2 + start_button.get_height() / 2)) #Website from above - this displays the start button text cenetered below the title 
        pygame.display.update() #Website from above - this updates the display to show any changes 

        while not self.start_game: #Website from above - this continues to loop until the game is started  
            for event in pygame.event.get(): #Website from above - this iterates through all pygame events
                if event.type == pygame.QUIT: #Website from above - this checks if the event type is a quit event  
                    pygame.quit() #Website from above - this quits the pygame module  
                    quit() #Website from above - this exits the program
                elif event.type == pygame.MOUSEBUTTONDOWN: #Website from above - this checks if the event is a mouse button down event 
                    self.start_game = True #Website from above - this starts the start_game function  

    def show_score(self): #Website from above - this defines the score showing on the screen or display 
        score_text = self.font.render('Score: ' + str(self.score), True, (255, 255, 255)) #Website from above - this renders the text with the current score using the font with white color
        self.screen.blit(score_text, (20, 10)) #Website from above - this displays the score text at a specific postion on the screen
        
    def game_over_screen(self): #Website from above - this defines the game over screen
        
        background_image = pygame.image.load('/Users/hajrahsiddiqui/Documents/cosmic cascade sounds/final.jpeg').convert_alpha()
        #background_image = pygame.image.load("/Users/akanshshah/Downloads/PHOTO-2024-05-28-16-27-05-6.jpg")
        self.screen.blit(background_image, (0, 0))
        
        #self.screen.fill((0, 0, 0)) #Website from above - this fill the screen with a black color  
        font = pygame.font.SysFont('Calibri', 1) #Website from above - this creates a font object with Calibri  
        title = font.render('Game Over!', True, (255, 255, 255)) #Website from above - this renders the text using the font with white color 
        restart_button = font.render('R - Restart', True, (255, 255, 255)) #Website from above - this renders the text using the font with white color 
        quit_button = font.render('Q - Quit', True, (255, 255, 255)) #Website from above - this renders the text using the font with white color 
        self.screen.blit(title, (self.screen_width / 2 - title.get_width() / 2, self.screen_height / 2 - title.get_height() / 3)) #GPT - this creates a display of the title text which is centered above the middle of the screen 
        self.screen.blit(restart_button, (self.screen_width / 2 - restart_button.get_width() / 2, self.screen_height / 1.9 + restart_button.get_height())) #GPT - this displays the restart button which is centered below the game over message  
        self.screen.blit(quit_button, (self.screen_width / 2 - quit_button.get_width() / 2, self.screen_height / 2 + quit_button.get_height() / 2)) #GPT - this displays the quit button which is centered below the title text 
        pygame.display.update() #Website from above - this updates the display to show any changes 

    def main(self): #Website from above - this is defining the main loop of the game  
        while not self.start_game: #Website from above - this continues the looping until the game is started 
            self.display_screen() #Website from above - this calls the method to display the game screen

        while not self.gameover: #Website from above - this continues the looping until the game is over 
            self.screen.fill((0, 0, 0)) #Website from above - this fills the screen with black color 
            self.draw_grid() #Website from above - this draws the grid on the screen
            self.draw_field() #Website from above - this draws the field on the screen
            self.draw_tile(self.current_tile) #Website from above - this draws the current tile on the screen
            self.draw_next_tile() #Website from above - this draws the next tile on the screen
            self.show_score() #Website from above - this draws the score on the screen
            pygame.display.flip() #Website from above - this updates the entire display to show any changes  
            self.clock.tick(60) #Website from above - this limits the frame rate to 60  

            self.fall_time += self.clock.get_rawtime() #GPT - this updates the fall time by adding the passed time 
            self.clock.tick() #Website from above - this ticks the clock to control the frame rate 
            if self.fall_time >= self.fall_speed: #Website from above - this checks if the fall time has exceeded the fall speed 
                self.fall_time = 0 #Website from above - this resets the fall time to zero
                self.shape_going_down() #Website from above - this allows the method to moe the current tile downwards 

            self.handle_events() #Website from above - this calls the method to handle the events for the current objects 

        self.game_over_screen() #Website from above - this displays the game over screen 
        while True:
            for event in pygame.event.get(): #Website from above - this iterates through all events that pygame has detected  
                if event.type == pygame.QUIT: #Website from above - this checks if the event type is a quit event, like for example, if the player wants to quit the game
                    pygame.quit() #Website from above - this quits the game window   
                    quit() #Website from above - this exits the program   
                elif event.type == pygame.KEYDOWN: #Website from above - this checks if the event type is a key press event
                    if event.key == pygame.K_r: #Website from above - the checks if the pressed key is the r key 
                        self.reset_game() #Website from above - this resets the game if the r is pressed by the player 
                        self.main() #Website from above - this ends the main loop function    
                        return  
                    elif event.key == pygame.K_q: #Website from above - this checks if the pressed key is the q key 
                        pygame.quit() #Website from above - this quits the game if the q is pressed by the player   
                        quit() #Website from above - this exits the program  

        def reset_game(self): #GPT - this resets the game state to its initial state 
            self.field = [[0 for y in range(W)] for x in range(H)] #GPT - this creates a 2D list representing the game field with their dimensions 
            self.score = 0 #Website from above - this initalizes the score to zero at the start of the game   
            self.gameover = False #Website from above - this indicates whether the game is over or not 
            self.current_tile = self.create_tile() #Website from above - this creates the current tile for the game 
            self.next_tile = self.create_tile() #Website from above - this creates the next tile for the game 
            self.fall_time = 0 #Website from above - this initializes the fall time to zero   
            self.start_game = False #Website from above - this tracks whether the game has started or not 

        def create_tile(self): #Website from above - this creates and returns a new tile at the top of the board 
            return Tile(5, 0, random.randint(0, len(shapes) - 1) % len(shape_colors)) #Website from above - this returns a new tile with a random shape or object 

        def draw_grid(self): #Website from above - this draws the grid lines of the game 
            for x in range(W): #Website from above - this iterates over the width of the grid 
                for y in range(H): #Website from above - this iterates over the height of the grid 
                    rect = pygame.Rect(x * TILE, y * TILE, TILE, TILE) #Website from above - this creates a rectangle for each tile in the grid 
                    pygame.draw.rect(self.screen, (40, 40, 40), rect, 1) #Website from above - this draws a rectangle on the screen with a boarder color 

        def intersect(self): #Website from above - this checks whether the current tile intersects with the boundries of the game field 
            intersection = False #Website from above - this initializes the intersection as false 
            for y in range(4): #Website from above - this loops through the rows of the current tile 
                for x in range(4): #Website from above - this loops through the colomns of the current tile  
                    if y * 4 + x in self.current_tile.recent_shape(): #Website from above - this checks whether the current shape intersects with the field 
                        if y + self.current_tile.y > H - 1: #Webite from above - this checks whether the tile has reached the bottom of the grid
                            return True 
                        if x + self.current_tile.x > W - 1: #Webite from above - this sets a condition where the tile has to reach the right edge of the grid
                            return True
                        if x + self.current_tile.x < 0: #Website from above - this sets a condition where the tile has to reach the left edge of the grid
                            return True
                        if self.field[y + self.current_tile.y][x + self.current_tile.x] > 0: #Website from above - this sets a condition where the tile has to intersect with a non empty cell in the field 
                            return True
            return False

        def break_lines(self): #Website from above - this checks for complete lines and then removes them, increasing the score later
            lines = 0 #Website from above - this initializes a variable to keep track of the number of lines that were removed 
            for y in range(1, H): #Website from above - this iterates over each row in the field 
                empty_cells = 0 #Website from above - this initializes a variable to count the empty cells in the row 
                for x in range(W): #Website from above - this iterates through each colomn in the row
                    if self.field[y][x] == 0: #Website from above - this checks if the cell in the current row is empty or not
                        empty_cells += 1 #Website from above - this increments the count of empty cells in the current row 
                if empty_cells == 0: #Website from above - this checks if the current row has no empty cells
                    lines += 1 #Website from above - this increments the number of lines cleared 
                    for y1 in range(y, 1, -1): #Website from above - this shifts the lines above the cleared lines
                        for x in range(W): #Website frome above - this continues to shift the lines 
                            self.field[y1][x] = self.field[y1 - 1][x] #Website from above - this moves the contents of the current row to the row above it 
            self.score += lines * 100 #Website from above - this increases the score by the number of lines cleared by 100 
            
            if self.score >= 200 and self.fall_speed > 75: #Website from above - this is basically just increasing the falling speed if the score is greater then or equal to 200 
                self.fall_speed = 75 #Website from above - this is basically just setting the falling speed to 70 frames per seconds
            if self.score >= 400 and self.fall_speed > 50: #Website from above - this is basically just increasing the falling speed if the score is greater then or equal to 400 
                self.fall_speed = 50 #Website from above - this is basically just setting the falling speed to 50 frames per seconds 
            if self.score >= 600 and self.fall_speed > 25: #Website from above - this is basically just increasing the falling speed if the score is greater than or equal to 600
                self.fall_speed = 25 #Website from above - this is basically just setting the falling speed to 25 frames per seconds 
            if self.score >= 800 and self.fall_speed > 10: #Website from above - this is basically just increasing the falling speed if the score if greater than or equal to 800
                self.fall_speed = 10 #Website from above - this is bascially just setting the falling speed to 10 frames per seconds
            if self.score >= 1000: #Website from above - this is basically just increasing the falling speed if the the score is greater then or equal to 1000 
                self.fall_speed = 1 #Website from above - this is basically just setting the falling speed to the lowest which is 1 frame per second

            self.next_tile = self.create_tile() #Website from above - this basically just generates a new tile object for the next tile
            if self.intersect(): #Website from above - this basically just checks if the current tile intersects with any other tile
                self.gameover = True #Website from above - this basically sets the game condition to true, if it meets the condition

        def draw_tile(self, tile): #Website from above - this draws the current falling tile on the screen 
            for y in range(4): #Website from above - this iterates over the rows of the following tiles
                for x in range(4): #Website from above - this iterates over the colomns of the following tiles 
                    if y * 4 + x in tile.recent_shape(): #Website from above - this checks if the x or y coordinate has a free tile or not
                        current_x = tile.x + x #Website from above - this calculates the current x coordinate on the grid
                        current_y = tile.y + y #Wesite from above - this calculates the current y coordinate on the grid 
                        if current_y >= 0: #Website from above - this checks whether the current y coordinate on the grid is visable in the grid
                            pygame.draw.rect(self.screen, shape_colors[tile.color], (current_x * TILE, current_y * TILE, TILE, TILE)) #Website from above - this draws the tile on the screen if the y coordinate is visable in the grid

        def draw_next_tile(self): #Website from above - this draws the preview of the next shape on the screen  
            label = self.font.render("Next Shape", True, (255, 255, 255)) #Website from above - this renders the label for the next shape preview  
            self.screen.blit(label, (game_width - DISPLAY_WIDTH + 20, 20)) #Website from above - this displays the label for the next shape preview on the screen
            format = self.next_tile.recent_shape() #Website from above - this gets the current shape format for the next tile
            for y in range(4): #Website from above - this loops through each row 
                for x in range(4): #Website from above - this loops through each coloumn 
                    if y * 4 + x in format: #Website from above - this checks if the current cell is part of the shape 
                        current_x = W * TILE + 50 + x * TILE #Website from above - this calculates the x coordinate of the current cell in the Next Shape 
                        current_y = 80 + y * TILE #Website from above - this calculates the y coordinate of the current cell in the Next Shape 
                        pygame.draw.rect(self.screen, shape_colors[self.next_tile.color], (current_x, current_y, TILE, TILE)) #Website from above - this draws a rectangle representing the Next Shape 

        def draw_field(self): #GPT - this draws the tiles that have already been fallen 
            for y in range(H): #Website from above - this iterates through each row in the field  
                for x in range(W): #Website from above - this iterates through each column in the field  
                    if self.field[y][x]: #Website from above - this checks if the cell is occupied or not 
                        pygame.draw.rect(self.screen, shape_colors[self.field[y][x] - 1], (x * TILE, y * TILE, TILE, TILE)) #Website from above - this draws the occupied with the shape color 

        def shape_going_down(self): #Website from above - this defines the shapes that are going down
            self.current_tile.y += 5 #Website from above - this moves the current tile down by one  
            if self.intersect(): #Website from above - this checks if the current tile intersects with any other tile 
                self.current_tile.y -= 5 #Website from above - this moves the current tile up by one  
                self.freeze() #Website from above - this locks the current tile in place  

        def freeze(self): #Website from above - this defines how the shapes get locked into their places on their grid
            for y in range(4): #Website from above - this loops through each row 
                for x in range(4): #Website from above - this loops through each column 
                    if y * 4 + x in self.current_tile.recent_shape(): #Website from above - this checks if the current position is part of the recent shape 
                        self.field[y + self.current_tile.y][x + self.current_tile.x] = self.current_tile.color + 1 #Website from above - this sets the field's cell at the current tile's position 
            self.break_lines() #Website from above - this calls the method to check and break any completed lines in the grid
            self.current_tile = self.next_tile #Website from above - this sets the current tile to the next tile  
            self.next_tile = self.create_tile() #Website from above - this geneartes a new tile to be used as the next tile in the game 
            if self.intersect(): #Website from above - this checks if the current tile intersects with any other tile  
                self.gameover = True #Website from above - this sets the gameover to condition to true  

        def go_horizontal(self, delta): #Website from above - this defines the method to move the current tile horizontally
            x = self.current_tile.x #Website from above - this stores the current x coordinate of the current tile  
            self.current_tile.x += delta #Website from above - this updates the x coordinate of the current tile  
            if self.intersect(): #Website from above - this checks if the current tile intersects with any other tile
                self.current_tile.x = x #Website from above - this resets the x coordinate of the current tile to its previous value

        def rotate(self): #Website from above - this defines the method to rotate the current tile 
            rotation = self.current_tile.rotation #Website from above - this stores the current rottion state of the current tile
            self.current_tile.shape_rotation() #Website from above - this rotates the shape of the current tile 
            if self.intersect(): #Website from above - this checks if the current tile intersects with any other tile 
                self.current_tile.rotation = rotation #Website from above - this resets the rotation state of the current tile to its previous value 

        def handle_events(self): #Website from above - this defines the possible events that a player could move or do 
            for event in pygame.event.get(): #Website from above - this itertes thorugh all pygame events  
                if event.type == pygame.QUIT: #Website from above - this checks if the event type is a quit event 
                    pygame.quit() #Website from above - this quits the pygame module
                    quit() #Website from above - this exits the program 
                if event.type == pygame.KEYDOWN: #Website from above - this checks if the event is a key press event  
                    if event.key == pygame.K_UP: #Website from above - this checks if the event is a key pressed up arrow
                        self.rotate() #Website from above - this rotates the current tile clockwise 
                    if event.key == pygame.K_DOWN: #Website from above - this checks if the event is a key pressed down arrow 
                        self.shape_going_down() #Website from above - this makes the current tile move downwards  
                    if event.key == pygame.K_RIGHT: #Website from above - this checks if the key pressed if the right arrow key  
                        self.go_horizontal(1) #Website from above - this moves the current tile on step to the right 
                    if event.key == pygame.K_LEFT: #Website from above - this checks if the key pressed is the left arrow key 
                        self.go_horizontal(-1) #Website from above - this moves the current tile one step to the left
                    if event.key == pygame.K_SPACE: #Website from above - this checks if the key pressed if the spacebar key 
                        while not self.intersect(): #Website from above - this continues the looping until the current tile intersects with any other tile 
                            self.current_tile.y += 1 #Website from above - this moves the current tile one step downwards  
                        self.current_tile.y -= 1 #Website from above - this  moves the current tile one step upwards 
                        self.freeze() #Website from above - this locks the current tile in place  

        def display_screen(self): #Website from above - this defines the display screen 
            
            background_image = pygame.image.load('/Users/hajrahsiddiqui/Documents/cosmic cascade sounds/no.png').convert_alpha()
            
            #background_image = pygame.image.load("/Users/akanshshah/Downloads/PHOTO-2024-05-28-16-27-05-6.jpg")
            self.screen.blit(background_image, (0, 0)) 
            
            #self.screen.fill((0, 0, 0)) #Website from above - this fills the screen with a black color 
            font = pygame.font.SysFont('Calibri', 1) #Website from above - this creates a font object with Calibri 
            title = font.render('Tetris Game!', True, (255, 255, 255)) #Website from above - this renders the text using the font with white color
            start_button = font.render('Click to start!', True, (255, 255, 255)) #Website from above - this renders the text using the font with white color 
            self.screen.blit(title, (self.screen.get_width() / 2 - title.get_width() / 2, self.screen.get_height() / 2 - title.get_height() / 2)) #Website from above - this displays the title text centered on the screen 
            self.screen.blit(start_button, (self.screen.get_width() / 2 - start_button.get_width() / 2, self.screen.get_height() / 2 + start_button.get_height() / 2)) #Website from above - this displays the start button text cenetered below the title 
            pygame.display.update() #Website from above - this updates the display to show any changes 

            while not self.start_game: #Website from above - this continues to loop until the game is started  
                for event in pygame.event.get(): #Website from above - this iterates through all pygame events
                    if event.type == pygame.QUIT: #Website from above - this checks if the event type is a quit event  
                        pygame.quit() #Website from above - this quits the pygame module  
                        quit() #Website from above - this exits the program
                    elif event.type == pygame.MOUSEBUTTONDOWN: #Website from above - this checks if the event is a mouse button down event 
                        self.start_game = True #Website from above - this starts the start_game function  

        def show_score(self): #Website from above - this defines the score showing on the screen or display 
            score_text = self.font.render('Score: ' + str(self.score), True, (255, 255, 255)) #Website from above - this renders the text with the current score using the font with white color
            self.screen.blit(score_text, (20, 10)) #Website from above - this displays the score text at a specific postion on the screen

        def game_over_screen(self): #Website from above - this defines the game over screen 
            
            background_image = pygame.image.load('/Users/hajrahsiddiqui/Documents/cosmic cascade sounds/final.jpeg').convert_alpha()
            #background_image = pygame.image.load("/Users/akanshshah/Downloads/PHOTO-2024-05-28-16-27-05-6.jpg")
            self.screen.blit(background_image, (0, 0)) 

            #self.screen.fill((0, 0, 0)) #Website from above - this fill the screen with a black color  
            font = pygame.font.SysFont('Calibri', 1) #Website from above - this creates a font object with Calibri  
            title = font.render('Game Over!', True, (255, 255, 255)) #Website from above - this renders the text using the font with white color 
            restart_button = font.render('R - Restart', True, (255, 255, 255)) #Website from above - this renders the text using the font with white color 
            quit_button = font.render('Q - Quit', True, (255, 255, 255)) #Website from above - this renders the text using the font with white color 
            self.screen.blit(title, (self.screen_width / 2 - title.get_width() / 2, self.screen_height / 2 - title.get_height() / 3)) #GPT - this creates a display of the title text which is centered above the middle of the screen 
            self.screen.blit(restart_button, (self.screen_width / 2 - restart_button.get_width() / 2, self.screen_height / 1.9 + restart_button.get_height())) #GPT - this displays the restart button which is centered below the game over message  
            self.screen.blit(quit_button, (self.screen_width / 2 - quit_button.get_width() / 2, self.screen_height / 2 + quit_button.get_height() / 2)) #GPT - this displays the quit button which is centered below the title text 
            pygame.display.update() #Website from above - this updates the display to show any changes 

        def main(self): #Website from above - this is defining the main loop of the game  
            while not self.start_game: #Website from above - this continues the looping until the game is started 
                self.display_screen() #Website from above - this calls the method to display the game screen

            while not self.gameover: #Website from above - this continues the looping until the game is over 
                self.screen.fill((0, 0, 0)) #Website from above - this fills the screen with black color 
                self.draw_grid() #Website from above - this draws the grid on the screen
                self.draw_field() #Website from above - this draws the field on the screen
                self.draw_tile(self.current_tile) #Website from above - this draws the current tile on the screen
                self.draw_next_tile() #Website from above - this draws the next tile on the screen
                self.show_score() #Website from above - this draws the score on the screen
                pygame.display.flip() #Website from above - this updates the entire display to show any changes  
                self.clock.tick(60) #Website from above - this limits the frame rate to 60  

                self.fall_time += self.clock.get_rawtime() #GPT - this updates the fall time by adding the passed time 
                self.clock.tick() #Website from above - this ticks the clock to control the frame rate 
                if self.fall_time >= self.fall_speed: #Website from above - this checks if the fall time has exceeded the fall speed 
                    self.fall_time = 0 #Website from above - this resets the fall time to zero
                    self.shape_going_down() #Website from above - this allows the method to moe the current tile downwards 

                self.handle_events() #Website from above - this calls the method to handle the events for the current objects 

            self.game_over_screen() #Website from above - this displays the game over screen 
            while True:
                for event in pygame.event.get(): #Website from above - this iterates through all events that pygame has detected  
                    if event.type == pygame.QUIT: #Website from above - this checks if the event type is a quit event, like for example, if the player wants to quit the game
                        pygame.quit() #Website from above - this quits the game window   
                        quit() #Website from above - this exits the program   
                    elif event.type == pygame.KEYDOWN: #Website from above - this checks if the event type is a key press event
                        if event.key == pygame.K_r: #Website from above - the checks if the pressed key is the r key 
                            self.reset_game() #Website from above - this resets the game if the r is pressed by the player 
                            self.main() #Website from above - this ends the main loop function    
                            return  
                        elif event.key == pygame.K_q: #Website from above - this checks if the pressed key is the q key 
                            pygame.quit() #Website from above - this quits the game if the q is pressed by the player   
                            quit() #Website from above - this exits the program  
            
if __name__ == "__main__": #Website from above - this checks if the program is being ran by python 
    game = Game() #Website from above - this creates an instance of the Game class to start the game  
    game.main() #Website from above - this calls the main method of the game  
```
